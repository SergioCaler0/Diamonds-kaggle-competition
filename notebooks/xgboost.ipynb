```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
from sklearn.model_selection import learning_curve, GridSearchCV
```

```python
df_train = pd.read_csv("../data/test-dumies.csv")
```

```python
df_train
```

```python
df_test = pd.read_csv("../data/test-dumies2p.csv")
```

```python
df_test
```

```python
X = df_train.drop("price", axis = 1)
y = df_train["price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.2, 
                                                    random_state=123)
```

```python
model = XGBRegressor()
params = {'nthread':[4], #when use hyperthread, xgboost may become slower
              'objective':['reg:linear'],
              'learning_rate': [0.02], #so called `eta` value
              'max_depth': [40],
              'min_child_weight': [1, 2],
              'silent': [1],
              'subsample': [0.7],
              'colsample_bytree': [0.7],
              'n_estimators': [1000]}
grid_search = GridSearchCV(model,
                           param_grid=params,
                           
                           
                           verbose=1)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_score_
```

```python
grid_search.best_estimator_
```

```python
best_XGB = grid_search.best_estimator_
best_XGB.score(X_test, y_test)
```

```python
best_XGB.fit(X, y)
```

```python
y_pred_train = best_XGB.predict(X_train)
y_pred_test = best_XGB.predict(X_test)
```

```python
mse_train = mean_squared_error(y_pred_train,y_train)
mse_test = mean_squared_error(y_pred_test,y_test)
```

```python
mse_train, mse_test
```

```python
prediction = best_XGB.predict(df_test)
```

```python
prediction
```

```python
test_XGBRegressor = pd.DataFrame(df_test["id"])
```

```python
test_XGBRegressor["price"] = prediction
```
